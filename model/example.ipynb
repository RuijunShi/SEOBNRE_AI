```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from generation_aiwavefrom import GenerationWaveform
waveform = GenerationWaveform()
```

```python
amp, pha = waveform.generation(np.array([2, 0.1, 0, 0]))
plt.figure(figsize=[10, 4])
merge_time = np.argmax(amp[0])
lem_wf = amp[0].shape[0]
time = np.linspace(0, lem_wf/4096, lem_wf) - merge_time/4096
plt.plot(time, amp[0] * np.exp(1j*pha[0].real) )
plt.grid()
plt.show()
```
