```python
import matplotlib.pyplot as plt
def log_loss(log_file_path):
    amp_loss_values = []
    test_loss_values = []

    with open(log_file_path, "r") as log_file:
        for line in log_file:
            line = line.split("- INFO - Epoch")
            if len(line) > 1:
                line = line[1]
                line_split = line.split("|")
                train_loss = line_split[1].split(":")[1]
                test_loss = line_split[2].split(":")[1]

                amp_loss_values.append(float(train_loss))
                test_loss_values.append(float(test_loss))
    print(f"train loss:{amp_loss_values[-1]}, test loss: {test_loss_values[-1]}")
    
    plt.figure(figsize=(8, 4), dpi=300)
    plt.semilogy(test_loss_values[:], label="Test Loss", color="green")
    plt.semilogy(amp_loss_values[:], label="Train Loss", color="orange")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
